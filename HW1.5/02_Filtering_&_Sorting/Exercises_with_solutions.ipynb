### Step 3. `chipo` 변수에 dataframe 저장하기

```python
chipo = pd.read_csv('./chipotle.tsv',sep='\t')
chipo
#urlretrieve('url-adress')
#https://seong6496.tistory.com/242
```

### Step 4. 가격(item_price)이 $10.00 이상인 데이터만 걸러내기

```python
# 새로운 변수에 item_price 열을 정리하고 float로 변환합니다
prices =  chipo['item_price'].str[1:].astype(float)
#https://seong6496.tistory.com/242
#https://rfriend.tistory.com/470

# chipo의 item_price를 위 prices 변수로 치환합니다
chipo.item_price = prices


# 'item_name', 'quantity', 'choice_description'의 중복을 제거하여 새로운 변수에 저장합니다.
chipo_filtered = chipo.drop_duplicates(subset=['item_name', 'quantity', 'choice_description'])
print(chipo_filtered)
# `chipo_filtered` 변수의 'quantity' 가 1인 데이터만 추려 새로운 변수에 저장합니다.
chipo_one_prod = chipo_filtered[chipo_filtered.quantity==1]
#chipo_one_prod


# 마지막으로 'chipo_one_prod'를 이용하여 item_price가 10달러 이상인 데이터를 걸러 봅니다 (위 과정은 중복을 피하기 위해 거친 단계입니다!)
chipo_one_prod=chipo_one_prod[chipo_one_prod.item_price>=10]


#아래 코드와 결과를 비교해보기
chipo_one_prod[chipo_one_prod.item_price>10].item_name.nunique()
```

### Step 5. 특정 칼럼만 선택하여 불러오기
#### `item_name` 과 `item_price` 칼럼 가져오기

```python
# 'item_name' and 'quantity' 칼럼의 중복을 제거하라!
chipo_filtered2 = chipo.drop_duplicates(['item_name', 'quantity'], ignore_index=True)
chipo[(chipo['item_name'] == 'Chicken Bowl') & (chipo['quantity'] == 1)]

# `quantity`가 1인 row만 가져와서 저장
chipo_one_prod = chipo_filtered2[chipo_filtered2.quantity ==1]
print(chipo_one_prod)

# 'item_name' 과 'item_price' 만 가져오기!
price_per_item = chipo[['item_name','item_price']]


# item_price을 이용하여 데이터 정렬해보기!!(Hint. sort_values)
price_per_item.sort_values('item_price')
```
