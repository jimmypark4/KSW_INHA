## <span style='color:black; background-color:#dcffe4'> Air Quality Dataset을 이용한 프로젝트 </span>
**"Air Quality"** dataset은 이탈리아의 한 도시에서 현장에 배치된 가스 멀티센서 장치의 응답을 포함하고 있습니다.

본 Dataset에는 대기질 화학 다중센서 장치에 내장된 5개의 금속 산화물 화학 센서에서 시간당 평균 응답의 9358개의 인스턴스가 포함되어 있습니다.

**누락된 값은 -200으로 입력되어있습니다.**

#### 실습 1
 
위 코드에서 볼 수 있듯이 현재 데이터에는 다양한 결측치가 있습니다. 

각 조별로 다양한 방법으로 결측치를 채워넣거나 삭제해보고, 데이터에서 확인할 수 있는 다양한 정보를 찾아봅시다!

```python
#1
#다양한 통계적 결측지 처리가 있습니다.  0으로 다바꿨습니다.
# https://choigww.github.io/data%20preprocessing/2021/02/18/%EB%8D%B0%EC%9D%B4%ED%84%B0-%EA%B2%B0%EC%B8%A1%EC%B9%98-%EC%B1%84%EC%9A%B0%EB%8A%94-6%EA%B0%80%EC%A7%80-%EB%B0%A9%EB%B2%95/
# Imputation Using (Mean/Median) Values:
# Imputation Using (Most Frequent) or (Zero/Constant) Values:
# Imputation Using k-NN:
# Imputation Using Deep Learning (Datawig):
import pandas as pd
import datetime 
import seaborn as sns
df = pd.read_csv('./AirQualityUCI.csv', sep=';')

df=df.fillna(0)
df.isnull().sum()

df = df.replace(-200,0)
df
```

#### 실습 2
 
위 데이터셋은 1시간 단위로 데이터를 수집했습니다.

하루 단위로 데이터를 묶어 봅시다! 이 때, 각 칼럼 값은 그 날 하루 칼럼 값의 평균으로 합니다.

하루 단위로 묶인 새로운 데이터셋의 유의미한 정보를 찾아봅시다! (예시, NO2 칼럼의 평균, 최대, 최솟값 등등)

```python
#2
#date를 인덱스로 한후 평균내서 만들었습니다.

df['Date']=pd.to_datetime(df['Date'])
df_day=df.set_index('Date')
df_day=df.groupby('Date').mean()
df_day.drop("1970-01-01")
df_day.min()
df_day.max()
```
